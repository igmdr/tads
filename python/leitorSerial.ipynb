```python
import serial

def criarSerial(portaSerial,baudRate):
    try:
        ser = serial.Serial(portaSerial,baudRate)
        print("Conectado ao Arduino porta serial: " + portaSerial)
        return ser
    except:
        print("Não foi possível conexão ao Arduino porta serial:" + portaSerial)
```

```python
from datetime import datetime

def criarNome():
    atual = datetime.now()
    nomeArq = atual.strftime("%H")+'h'+atual.strftime("%M")+'m'+atual.strftime("%f")+'ms'+".csv"
    return nomeArq
```

```python
from datetime import datetime

def retornarHorario():
    atual = datetime.now()
    horario = atual.strftime("%H")+atual.strftime("%M")+atual.strftime("%f")
    return horario
```

```python
def criarArquivo():
    nomeArq = criarNome()
    caminhoArq = open(nomeArq,'a')
    print("Arquivo criado")
    return nomeArq
```

```python
def gravarArquivo(nomeArq,dados):
    caminhoArq = open(nomeArq,'a')
    caminhoArq.write(dados + '; ' + "; ; ; ;" + retornarHorario() + "\n") #escreve os dados e pula de linha
    caminhoArq.close()#fecha o arquivo
```

```python
def fecharLeitura(ser):
    ser.close()
    print("Arquivo salvo e conexão serial encerrada")
```
